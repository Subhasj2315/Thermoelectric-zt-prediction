```python
import pandas as pd
df = pd.read_csv("Cleaned_Data.csv")
```

```python
df.head(10)
```

```python
from pymatgen.core import Composition

df["composition"] = df["Formula"].apply(Composition)
```

```python
df
```

```python
df["composition"]
```

```python
from matminer.featurizers.composition import ElementProperty

featurizer = ElementProperty.from_preset("magpie")
```

```python
df_features = featurizer.featurize_dataframe(
    df,
    col_id="composition",
    ignore_errors=True
)
```

```python
df_features
```

* So, here we are getting in total 132 vector features.

```python
df_features.shape
```

```python
df_features.columns
```

```python

X = df_features.drop(columns=["Formula","composition","ZT"]) # Removing no-numeric columns

Y = df_features["ZT"]
```

```python
X
```

```python
Y
```

```python
X.isnull().sum()
# no. of missing values in each columns
```

```python
missing = X.isnull().sum()

print(missing[missing == 25])
```

All the feature columns are having 25 missing values. So, that's why we will use median imputation as missing values are few as compared to total no. of 
rows.

```python
X = X.fillna(X.median())
```

```python
X.isnull().sum()
```

```python
Y.shape 
```

```python
X.shape 
```

```python
# save the feature matrix

feature_df = X.copy()
feature_df["ZT"] = Y

feature_df.to_csv(
    "Magpie_Features.csv",
    index=False
)
```

* so, there are in total 139 unique compounds and we have to split the test and train set by unique compounds.

```python
unique_compounds = df["Formula"].unique()
```

```python
len(unique_compounds)
```

```python
unique_compounds[:10]
```

```python
# now splitting of training and test set using Scikit learn

from sklearn.model_selection import train_test_split

train_compounds, test_compounds = train_test_split(
    unique_compounds,
    test_size=0.20,
    random_state=42
)
```

```python
print(len(train_compounds))
print(len(test_compounds))
```

```python
# Create Masks to get the actual rows for test and train compounds
train_mask = df["Formula"].isin(train_compounds)
test_mask = df["Formula"].isin(test_compounds)
#boolean arrays 
```

```python
X_train = X[train_mask]
Y_train = Y[train_mask]
X_test = X[test_mask]
Y_test = Y[test_mask]
```

```python
X_train
```

```python
X_test
```

```python
# to check the data leakage 
train_formulas = set(df.loc[train_mask,"Formula"])

test_formulas = set(df.loc[test_mask,"Formula"])

train_formulas.intersection(test_formulas)
```

So, this indicates the split of train and test is done without any leakage

```python
print(X_train.shape)
print(X_test.shape)

print(Y_train.shape)
print(Y_test.shape)
```

```python
X_train.to_csv("x_train.csv", index = False)
X_test.to_csv("x_test.csv",index = False)
Y_test.to_csv("y_test.csv",index = False)
Y_train.to_csv("y_train.csv",index = False)
```
